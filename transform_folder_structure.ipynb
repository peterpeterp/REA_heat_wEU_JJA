```python
for step in range(12,-1,-1):
    # identify case that is used as source for precompiled copy
    directories = sorted(glob.glob(f"{exp.dir_archive}/GKLT/{exp.experiment_name}{'/*' * (step+1)}/atm"))
    directories = [d.replace('/atm','') for d in directories]
    directories = [d for d in directories if 'step' not in d.split('/')[-2]]
    if len(directories) == 100:
        os.system(f'rm -rf {exp.dir_archive}/GKLT/{exp.experiment_name}/step{step}')
        for source in directories:
            print(source)
            destination = f"{exp.dir_archive}/GKLT/{exp.experiment_name}/step{step}/{source.split('/')[-1]}"
            os.makedirs(destination, exist_ok=True)
            subdirs = [s.split('/')[-1] for s in glob.glob(f"{source}/*")]
            for subdir in [s.split('/')[-1] for s in glob.glob(f"{source}/*")]:
                if subdir != 'rest' and exp.experiment_identifier not in subdir:
                    _ = os.system(f"cp -r {source}/{subdir} {destination}/{subdir}")
```
